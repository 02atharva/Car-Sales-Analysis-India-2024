```python
# in feb column the data  present as 2,11 (object)instead 211 (int) change the datatype and remove " , "from the cell

df2['February'] = pd.to_numeric(df2['February'].astype(str).str.replace(',', ''), errors='coerce')
```

```python
#replacing the missed june value to zero
df2['June'] = df2['June'].fillna('0')
# change datatype
df2["June"] = df2["June"].astype('int64')
```

```python
df2['MoM %'] = df2['MoM %'].astype('float')
df2['YoY %'] = df2['YoY %'].astype('float')
df2.info()
```

```python
df1['MoM %'] = df1['MoM %'].astype('float')
df1['YoY %'] = df1['YoY %'].astype('float')
df1.info()
```


<font size="5">Explore Data Analysis</font>


```python
df1.head(12)
```

```python
sns.set_style("whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)
```


<font size="5">1) Sales by Car Segment -</font>


```python
# Segment-wise total sales
seg_sales = (
    df1.groupby('Segment', as_index=False)['Sales']
    .sum()
    .sort_values('Sales', ascending=False)
)

plt.figure(figsize=(10, 6))
sns.barplot(
    x='Segment',
    y='Sales',
    data=seg_sales,
    hue='Segment',
    palette='viridis',
    legend=False
)

plt.title('Total Sales by Segment', fontweight='bold')
plt.xlabel('Segment')
plt.ylabel('Total Sales')
plt.tight_layout()
plt.show()
```

<font size="5">2) Monthly Sales Trend -</font>

```python
# CLEAN Months column BEFORE making it categorical
df1['Months'] = (
    df1['Months']
    .astype(str)
    .str.strip()
    .str.title()
)
```

```python
monthly_sales = df1.groupby(
    'Months',
    as_index=False,
    observed=True
)['Sales'].sum()

sns.lineplot(
    x='Months',
    y='Sales',
    data=monthly_sales,
    marker='o',
    linewidth=2
)
plt.title("Monthly Sales Trend", fontweight='bold')
plt.xticks(rotation=30)
plt.show()
```


<font size="5">3) Distribution of Sales -</font>


```python
plt.figure(figsize=(10, 6))
sns.histplot(df1['Sales'], bins=20, kde=True)

plt.title('Distribution of Sales', fontweight='bold')
plt.xlabel('Sales')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```


<font size="5">4) Top 10 Cars by Sales -</font>


```python
top_models = (df1.groupby('Model')['Sales'].sum().sort_values(ascending=False).head(10).reset_index())

sns.barplot(x='Sales', y='Model', data=top_models, palette='Blues_r', hue='Model')
plt.title("Top 10 Cars by Sales")
plt.ylabel("Cars")
plt.xlabel("Sales of Car Units")
plt.show()
```

<font size="5">5) Bottom 10 Cars by Sales -</font>

```python
bottom_10_models = (
    df1.groupby('Model', as_index=False)['Sales']
    .sum()
    .sort_values('Sales', ascending=True)
    .head(10)
)

plt.figure(figsize=(10, 6))
sns.barplot(
    x='Sales',
    y='Model',
    data=bottom_10_models,
    palette='Reds',
    hue='Model'
)

plt.title('Bottom 10 Models by Sales', fontweight='bold')
plt.xlabel('Total Sales')
plt.ylabel('Car Model')
plt.tight_layout()
plt.show()
```

<font size="5">6) Market Share by Body Type -</font>


```python
make_sales = (
    df1.groupby('Make')['Sales']
    .sum()
    .sort_values(ascending=False)
)

top_make_sales = make_sales.head(7)

plt.figure(figsize=(7, 7))
plt.pie(
    top_make_sales,
    labels=top_make_sales.index,
    autopct='%1.1f%%',
    startangle=90,
    counterclock=False
)

plt.title('Market Share by Top 7 Brands', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```


<font size="5">7) Average Sales of Models per Body Type - </font>


```python
avg_sales_body = (
    df1.groupby('Body Type', as_index=False)['Sales']
    .mean()
    .sort_values('Sales', ascending=False)
)

plt.figure(figsize=(10, 6))
sns.barplot(
    x='Body Type',
    y='Sales',
    data=avg_sales_body,
    palette='Set2',
    hue='Body Type'
)

plt.title('Average Sales by Body Type', fontweight='bold')
plt.xlabel('Body Type')
plt.ylabel('Average Sales')
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
```
